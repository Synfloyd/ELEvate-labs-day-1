# 2. How do you treat duplicate records ?

In Pandas, I identify and analyze duplicate records, then remove them if unnecessary. I ensure data integrity by keeping relevant entries and dropping exact or column-specific duplicates as needed.

# 3. Difference between dropna and fillna in pandas ?

In Pandas, dropna() is used to remove rows or columns with missing values, while fillna() is used to replace missing values with a specified value, like mean or median. dropna() reduces dataset size, whereas fillna() retains structure and is useful for maintaining data consistency in analysis.

# 4. What is outlier treatment and why it is important ?

Outlier treatment involves identifying and handling extreme values that deviate significantly from other data points. It's important because outliers can skew analysis, distort statistical measures, and impact model accuracy. Proper treatment—like removal, transformation, or capping—ensures data quality, improves insights, and leads to more reliable and stable analytical results.

# 5. Explain the process of standardizing data ?

Data standardization converts data into a standard format that computers can read and understand. This is important because it allows different systems to share and efficiently use data. Without data standardization, it would not be effortless for different approaches to communicate and exchange information.

Data standardization is also essential for preserving data quality. When data is standardized, it is much easier to detect errors and ensure that it is accurate. This is essential for making sure that decision-makers have access to accurate and reliable information.

# 7. what are common data cleaning challenges ?

Missing values – Incomplete data can affect accuracy and model performance.
Duplicate records – Redundant rows can skew results and waste storage.
Inconsistent formats – Date, time, currency, or text format variations make data hard to analyze.
Outliers – Extreme values may distort analysis if not treated properly.
Typographical errors – Spelling mistakes or inconsistent labels in categorical data.
Irrelevant data – Unnecessary columns or outdated information that add noise.
Mixed data types – Columns with incorrect or inconsistent data types.
Encoding issues – Special characters or non-UTF-8 text causing reading errors.
Unit inconsistency – Mixing units (e.g., kg and lbs) in the same column.
Hidden nulls – Empty strings or placeholders like “N/A” that aren’t recognized as missing.

# 8. How can you check data quality ?

Completeness - Check for missing values using .isnull().sum().

Accuracy - Validate values against known rules or sources (e.g., age can’t be negative).

Consistency - Look for uniform formats, naming conventions, and values (e.g., "Male" vs "male").

Uniqueness - Check for duplicate rows or IDs.

Validity - Ensure data matches expected formats or ranges (e.g., dates, numeric limits).

Timeliness - Confirm data is up to date and relevant.
